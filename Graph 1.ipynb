```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_names'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Count of records from each class using crosstab
class_counts = pd.crosstab(index=df['target'], columns='count')
print(class_counts)

# Histogram plot for each feature
df.hist(bins=20, figsize=(10, 8))
plt.suptitle('Histogram of Iris Dataset Features')
plt.show()

# Box plot for each feature grouped by target
plt.figure(figsize=(10, 6))
df_melt = pd.melt(df, id_vars=['target_names'], value_vars=iris.feature_names)
sns.boxplot(x='variable', y='value', hue='target_names', data=df_melt)
plt.title('Box Plot of Iris Dataset Features')
plt.show()

# Scatter plot for the first two features
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target_names', data=df)
plt.title('Scatter Plot of Sepal Length and Width')
plt.show()

# Scatter plot for petal features
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target_names', data=df)
plt.title('Scatter Plot of Petal Length and Width')
plt.show()
```
